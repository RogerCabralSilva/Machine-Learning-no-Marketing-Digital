# Importação

```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")

df
```

# Arrumando as colunas json

```python
import json

# colunas no formato json no df
json_columns = ["device", "geoNetwork", "totals", "trafficSource"]

# Converter cada coluna JSON de string para dicionário
for col in json_columns:
    df = df.join(pd.DataFrame([json.loads(linha) for linha in df[col]]))
    
df.drop(columns=json_columns, inplace=True)

df
```

# Mudando o tipo das colunas

```python
df.dtypes
```

```python
df[["date", "fullVisitorId", "visitId"]] = df[["date", "fullVisitorId", "visitId"]].astype(object)

df.dtypes
```
